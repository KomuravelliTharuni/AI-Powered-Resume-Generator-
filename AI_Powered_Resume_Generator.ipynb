```python
import gradio as gr
from transformers import pipeline
from reportlab.lib.pagesizes import A4
from reportlab.pdfgen import canvas
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication
from email.mime.text import MIMEText
import smtplib
import os
import datetime

# Load open-source LLM
generator = pipeline("text-generation", model="tiiuae/falcon-rw-1b")

# Generate resume content
def generate_resume_text(name, email, phone, skills, tools, education, project):
    prompt = f"""
    Create a modern and professional resume for:
    Name: {name}
    Phone: {phone}
    Email: {email}
    Skills: {skills}
    Tools: {tools}
    Education: {education}
    Projects: {project}

    Focus on Data Analyst role with 3 years of experience.
    """
    result = generator(prompt, max_length=500, do_sample=True)
    return result[0]['generated_text']

# Create PDF with visual design
def create_resume_pdf(name, resume_text):
    filename = f"{name.replace(' ', '_')}_Resume.pdf"
    filepath = os.path.join(os.getcwd(), filename)
    c = canvas.Canvas(filepath, pagesize=A4)

    width, height = A4
    c.setFont("Helvetica-Bold", 20)
    c.setFillColorRGB(0.2, 0.4, 0.6)
    c.drawCentredString(width / 2.0, height - 80, f"{name}'s Resume")

    c.setFont("Helvetica", 12)
    c.setFillColorRGB(0, 0, 0)

    lines = resume_text.split('\n')
    y = height - 120
    for line in lines:
        if y < 50:
            c.showPage()
            y = height - 50
        c.drawString(50, y, line.strip())
        y -= 20

    c.save()
    return filepath

# Email resume
def send_email(pdf_path, receiver_email):
    sender_email = "tharunikomuravelli@gmail.com"
    password = "aqux eesd pqvn qdpg"

    msg = MIMEMultipart()
    msg["Subject"] = "Your Auto-Generated Resume"
    msg["From"] = sender_email
    msg["To"] = receiver_email

    body = MIMEText("Hi,\n\nAttached is your resume generated by AI.\n\nBest regards,\nResume Bot")
    msg.attach(body)

    with open(pdf_path, "rb") as f:
        part = MIMEApplication(f.read(), Name=os.path.basename(pdf_path))
        part['Content-Disposition'] = f'attachment; filename="{os.path.basename(pdf_path)}"'
        msg.attach(part)

    try:
        with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
            server.login(sender_email, password)
            server.send_message(msg)
        return "✅ Email sent successfully!"
    except Exception as e:
        return f"❌ Failed to send email: {str(e)}"

# Final wrapper function
def agentic_resume(name, email, phone, skills, tools, education, project):
    generated_text = generate_resume_text(name, email, phone, skills, tools, education, project)
    pdf_path = create_resume_pdf(name, generated_text)
    status = send_email(pdf_path, email)
    return f"Resume generated and saved as:\n{pdf_path}\n\n{status}", pdf_path

# Gradio Interface
with gr.Blocks() as demo:
    gr.Markdown("## 🤖 AI-Powered Resume Generator (Agentic)")

    with gr.Row():
        name = gr.Textbox(label="Full Name")
        email = gr.Textbox(label="Your Email")
        phone = gr.Textbox(label="Phone Number")

    with gr.Row():
        skills = gr.Textbox(label="Top Skills")
        tools = gr.Textbox(label="Tools & Technologies")

    education = gr.Textbox(label="Education")
    project = gr.Textbox(label="Key Project")

    generate_button = gr.Button("Generate My Resume 🎯")
    output = gr.Textbox(label="Status / Info")
    file_output = gr.File(label="Download Resume PDF")

    generate_button.click(agentic_resume,
        inputs=[name, email, phone, skills, tools, education, project],
        outputs=[output, file_output]
    )

demo.launch()
```
